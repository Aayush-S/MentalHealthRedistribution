```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alcohol_use = pd.read_csv('../../Platform/csv/Alcohol_use_disorders.csv')
drug_use = pd.read_csv('../../Platform/csv/Drug_use_disorders.csv')
violence = pd.read_csv('../../Platform/csv/Interpersonal_violence.csv')
self_harm = pd.read_csv('../../Platform/csv/Self-harm.csv')
```

```python
alcohol_use.tail()
```

```python
# remove first and last 2 rows
def removeRows(df):
    df.columns = df.iloc[0]
    df = df[2:-2]
    return df
alcohol_use = removeRows(alcohol_use)
drug_use = removeRows(drug_use)
violence = removeRows(violence)
self_harm = removeRows(self_harm)
```
